```python
import keras
```

```python
from keras import applications
from keras.preprocessing.image import ImageDataGenerator
from keras import optimizers
from keras.models import Sequential
from keras.layers import Dropout, Flatten, Dense, Conv2D
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
keras.backend.clear_session()
```

```python
# Load the VGG16 network
input_shape = (64, 64, 3)
base_model = applications.VGG16(weights='imagenet', 
                                include_top=False, input_shape=input_shape)
```

```python
# Create a new model
model = Sequential()
# Loop over base_model.layers and add each layer to model
for layer in base_model.layers:
    model.add(layer)
for layer in model.layers:
    layer.trainable=False
```

```python
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
train_data_dir = './train'
batch_size = 20
train_datagen = ImageDataGenerator(rescale=1./255,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True)
train_generator = train_datagen.flow_from_directory(
                        train_data_dir,
                        target_size=(64,64),
                        batch_size=batch_size,
                        class_mode='binary')
```

```python
test_data_dir = './test'
batch_size = 1
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
                        test_data_dir,
                        target_size=(512,512),
                        batch_size=batch_size,
                        class_mode='binary')
```

```python
# Display the image
def disp_image(im):
    if (len(im.shape) == 2):
        # Gray scale image
        plt.imshow(im, cmap='gray')    
    else:
        # Color image.  
        im1 = (im-np.min(im))/(np.max(im)-np.min(im))*255
        im1 = im1.astype(np.uint8)
        plt.imshow(im1)    
        
    # Remove axis ticks
    plt.xticks([])
    plt.yticks([])
```

```python
X = train_generator.next()

plt.figure(figsize=(10,10))
nplot = 4
for i in range(nplot):    
    plt.subplot(1,nplot,i+1)
    disp_image(X[0][i,:,:,:])
    if X[1][i].astype(int)==1:
        plt.xlabel('nonface')
    elif X[1][i].astype(int)==0:
        plt.xlabel('face')
```

```python
model.compile(optimizer=optimizers.Adam(lr=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x=X[0], y=X[1])
```

```python
import time

probs = []
locations = []

start = time.time()
for row in range(112):
    for col in range(112):
        pic = Xtest[0][:, row*4:row*4+64, col*4:col*4+64, :]
        probability = model.predict(pic)
        locations.append([row, col])
        probs.append(probability)
    if row%2==0:
        print('In process-------row number:',row)
end = time.time()
print('Total time:'+str((end-start))+'s')
```

```python
def draw_box(image, probability, location, stride):
    draw = ImageDraw.Draw(image)
    top = (location[0]*stride).astype('int32')
    left = (location[1]*stride).astype('int32')
    bottom = (location[0]*stride+64).astype('int32')
    right = (location[1]*stride+64).astype('int32')
    draw.rectangle([left, top, right, bottom])
```
